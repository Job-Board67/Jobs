```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Evan"],
    "Age": [19, 22, 21, 20, 23],
    "Score": [78, 85, 92, 88, 75]
}

df = pd.DataFrame(data)
print(df)

# Students older than 20
older_than_20 = df[df["Age"] > 20]
print("Students older than 20:")
print(older_than_20)

# Students with score above 80
score_above_80 = df[df["Score"] > 80]
print("Students with score above 80:")
print(score_above_80)
```
