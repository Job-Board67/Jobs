```python
import matplotlib.pyplot as plt

names = ["Alice", "Bob", "Charlie", "Diana", "Evan"]
scores = [78, 85, 92, 88, 75]
plt.plot(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

names = ["Alice", "Bob", "Charlie", "Diana", "Evan"]
scores = [78, 85, 92, 88, 75]
plt.bar(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```
